```python
from IPython.core.formatters import JavascriptFormatter
# oxoxoxox
# xoxoxoxo
# oxoxoxox
# 10 * 8

symbols =["o", "x"]

for i in range(10):
    line =""
    for j in range(8):
        # 開始位置不同 i=1 s=0, i=1 s=1, i=2 s=0
        symbols_pos = (i+j) % len(symbols)
        line = line + symbols[symbols_pos]
    print(line)
```
