```python
def is_leap_year(year):
    ans = False
    if year % 400 == 0:
        ans = True
    elif year % 100 == 0:
        ans = False
    elif year % 4 == 0:
        ans = True
    else:
        ans = False
    return ans
```

```python
def is_date_valid(date_checked):
    day = date_checked % 100
    remain = date_checked // 100
    month = remain % 100
    year = remain // 100

    if year > 0:
        if 1 <= month <= 12:
            day_limit_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_limit = day_limit_per_month[month-1]
            if month == 2 and is_leap_year(year) == True:
                day_limit = day_limit + 1
            if 1 <= day <= day_limit:
                # 日期合理
                return True
    # 預設不是對的
    return False
```

```python
def fully_squared_1(start=4499, max_count=10):
    # start = 4499
    ans = []
    count = 0
    while True:
        date_checked = start ** 2
        if is_date_valid(date_checked) == True:
            # print(date_checked)
            ans.append(date_checked)
            count = count + 1
        if count == max_count:
            break
        start = start + 1
    return ans
# fully_squared_1()
```

```python
import datetime as dt

def fully_squared_2(start_y=2024, start_m=10, start_d=1, max_count=10):
    ans = []
    pat = "%Y%m%d"
    start = dt.datetime(start_y, start_m, start_d)
    count = 0
    while True:
        start_str = start.strftime(pat)
        start_int = int(start_str)
        start_root = start_int ** 0.5
        # 四捨五入-原始 < 某個數值 我就認為依樣
        dif = abs(round(start_root) - start_root)
        if dif < 0.000001:
            # print(start)
            ans.append(start_int)
            count = count + 1
        if count == max_count:
            break
        start = start + dt.timedelta(days=1)
    return ans
# fully_squared_2()
```

```python
# 想法1(_2)和想法2(_1)到底差多少
import time

try_count = 50

start = time.time()
fully_squared_1(max_count=try_count)
end = time.time()
dif1 = end - start
print("測試1(想法2):", dif1, "秒")

start = time.time()
fully_squared_2(max_count=try_count)
end = time.time()
dif2 = end - start
print("測試2(想法1):", dif2, "秒")

ratio = dif2 / dif1
print("想法1比想法2慢了幾倍:", ratio)
```
