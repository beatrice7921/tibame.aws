```python
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# print(type(content))
content = json.loads(content)
# print(type(content))

f = open("danmu.csv","w",encoding="utf-8")
f.write("uid,text\n")
denmus = content["data"]["danmu"]
for d in denmus:
    d_uid = d["userid"]
    d_text = d["text"]
    pat ='{},{}'"\n"
    f.write(pat.format(d_uid,d_text))
f.close()

    # print(d_uid,d_text)
```
